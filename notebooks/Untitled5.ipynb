```python
from sklearn.feature_extraction.text import TfidfVectorizer
from os import listdir
from helper import *
```

```python
def get_iter_of_filenames(directory):
    return iter(list(map(lambda x: "{}/{}".format(directory, x), listdir(directory))))

contents = list(map(lambda x: x.replace("\n", "").replace(".", ""), list(transcripts_to_df("test_transcripts")["content"])))
```

```python
contents[0:2]
```

```python
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(contents)
```

```python
print(vectorizer.get_feature_names())
```

```python
X.shape
```

```python
import numpy as np
feature_names = np.array(vectorizer.get_feature_names())
responses = vectorizer.transform(contents[0:5])


def get_top_tf_idf_words(response, top_n=100):
    sorted_nzs = np.argsort(response.data)[:-(top_n+1):-1]
    return feature_names[response.indices[sorted_nzs]]
  
print([get_top_tf_idf_words(response,20) for response in responses])
```
